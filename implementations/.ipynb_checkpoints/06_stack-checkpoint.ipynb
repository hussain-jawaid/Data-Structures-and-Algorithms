```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        self.n = 0

    def get_length(self):
        return self.n

    def push(self, data):
        new_node = Node(data)
        if self.is_empty():
            self.top = new_node
            self.n += 1
            return
        new_node.next = self.top
        self.top = new_node
        self.n += 1

    def pop(self):
        if self.is_empty():
            raise Exception("Stack is empty")
        value = self.top.data
        self.top = self.top.next
        self.n -= 1
        return value

    def peek(self):
        if self.is_empty():
            raise Exception("Stack is empty")
        return self.top.data

    def search(self, value):
        if self.is_empty():
            raise Exception("Stack is empty")
        current_node = self.top
        index = 0
        while current_node:
            if current_node.data == value:
                return index
            current_node = current_node.next
            index += 1
        raise ValueError("Value not found")

    def is_empty(self):
        return self.top is None
```

```python
s = Stack()
```

```python
s.get_length()
```

```python
s.push(1)
s.push(2)
s.push(3)
```

```python
s.get_length()
```

```python
s.pop()
```

```python
s.peek()
```

```python
s.search(2)
```

### Problem
Given a string made up of the following brackets: ()[]{}, determine whether the brackets properly match.

```python
def brackets_match(expr):
    s = Stack()
    pairs = {
        "(": ")",
        "{": "}",
        "[": "]",
    }
    for char in expr:
        if char in pairs:
            s.push(char)
        elif char in pairs.values():
            if s.is_empty() or pairs[s.pop()] != char: return False
    return s.is_empty()
```

```python
expr = "aa"
brackets_match(expr)
```
